# Plotting Data and Vizualization

### Basic plots examples

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Line Plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure()
plt.plot(x, y)
plt.title('Line Plot')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()

# 2. Scatter Plot
x = np.random.rand(50)
y = np.random.rand(50)
plt.figure()
plt.scatter(x, y, c='blue')
plt.title('Scatter Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

x = np.random.rand(50)
y = np.random.rand(50)
s =np.random.rand(50) * 100  # Size of points
plt.figure()
plt.scatter(x, y, s, c='blue', alpha=s/100)
plt.title('Scatter Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# 3. Bar Plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 4]
plt.figure()
plt.barh(categories, values)
plt.title('Bar Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()

# 4. Bar Plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 4]
plt.figure()
plt.bar(categories, values)
plt.title('Bar Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()

# 5. Histogram
data = np.random.randn(1000)
plt.figure()
plt.hist(data, bins=30, color='purple', alpha=0.7)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# 6. Pie Chart
sizes = [15, 30, 45, 10]
labels = ['Apple', 'Banana', 'Cherry', 'Date']
plt.figure()
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()

# 7. Box Plot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.figure()
plt.boxplot(data)
plt.title('Box Plot')
plt.xlabel('Group')
plt.ylabel('Value')
plt.show()

# 8. Area Plot (Stacked)
x = np.arange(0, 10, 1)
y1 = np.random.rand(10)
y2 = np.random.rand(10)
plt.figure()
plt.stackplot(x, y1, y2, labels=['y1', 'y2'])
plt.title('Stacked Area Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

# 9. Heatmap
data = np.random.rand(10, 10)
plt.figure()
plt.imshow(data, cmap='viridis', aspect='auto')
plt.title('Heatmap')
plt.colorbar()
plt.show()

# 10. Error Bar Plot
x = np.linspace(0, 10, 10)
y = np.sin(x)
errors = 0.1 + 0.2 * np.sqrt(x)
plt.figure()
plt.errorbar(x, y, yerr=errors, fmt='o')
plt.title('Error Bar Plot')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()

# 11. Violin Plot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.figure()
plt.violinplot(data)
plt.title('Violin Plot')
plt.xlabel('Group')
plt.ylabel('Value')
plt.show()
```

### Advanced Plotting with extra libraries
requires installation of seaborn: `pip install seaborn`

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)
x = np.linspace(0, 10, 100)
y = 2 * x + 1 + np.random.normal(scale=2, size=100)
df = pd.DataFrame({'x': x, 'y': y})

# Simple scatter plot with linear regression fit and confidence interval
plt.figure()
sns.regplot(x='x', y='y', data=df, ci=95, line_kws={'color': 'red'})
plt.title('Seaborn Scatter Plot with Linear Fit and 95% CI')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Scatter plot with quadratic (order=2) regression fit and confidence interval
plt.figure()
sns.regplot(x='x', y='y', data=df, order=2, ci=95, line_kws={'color': 'green'})
plt.title('Seaborn Scatter Plot with Quadratic Fit and 95% CI')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Scatter plot with higher order regression fit and confidence interval
plt.figure()
sns.regplot(x='x', y='y', data=df, order=10, ci=95, line_kws={'color': 'blue'})
plt.title('Seaborn Scatter Plot with LOWESS Fit and 95% CI')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Some specific data representation might prove hard to represent without the use of extra libraries, consider wanting to create a heat map basedon an actual map of the state in the USA.
You would need to provide map as background and provide a pixle coordinate map to overlay for color fill boundaries
instead use some library that can do that for you, in our case for this example we use plotly: `pip install plotly` 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Define US states and abbreviations
states = [
    'Alabama', 'Alaska', 'Arizona', 'Arkansas', 'California', 'Colorado', 'Connecticut', 'Delaware',
    'Florida', 'Georgia', 'Hawaii', 'Idaho', 'Illinois', 'Indiana', 'Iowa', 'Kansas', 'Kentucky',
    'Louisiana', 'Maine', 'Maryland', 'Massachusetts', 'Michigan', 'Minnesota', 'Mississippi',
    'Missouri', 'Montana', 'Nebraska', 'Nevada', 'New Hampshire', 'New Jersey', 'New Mexico',
    'New York', 'North Carolina', 'North Dakota', 'Ohio', 'Oklahoma', 'Oregon', 'Pennsylvania',
    'Rhode Island', 'South Carolina', 'South Dakota', 'Tennessee', 'Texas', 'Utah', 'Vermont',
    'Virginia', 'Washington', 'West Virginia', 'Wisconsin', 'Wyoming'
]
state_abbr = [
    'AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE',
    'FL', 'GA', 'HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY',
    'LA', 'ME', 'MD', 'MA', 'MI', 'MN', 'MS', 'MO', 'MT',
    'NE', 'NV', 'NH', 'NJ', 'NM', 'NY', 'NC', 'ND', 'OH',
    'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX', 'UT',
    'VT', 'VA', 'WA', 'WV', 'WI', 'WY'
]

# Define which states are considered "East" and "West"
east_states = {
    'Maine', 'New Hampshire', 'Vermont', 'Massachusetts', 'Rhode Island', 'Connecticut', 'New York',
    'New Jersey', 'Pennsylvania', 'Delaware', 'Maryland', 'Virginia', 'West Virginia', 'North Carolina',
    'South Carolina', 'Georgia', 'Florida', 'Ohio', 'Indiana', 'Illinois', 'Michigan', 'Wisconsin',
    'Kentucky', 'Tennessee', 'Alabama', 'Mississippi'
}
# All others are "West"

np.random.seed(42)
results = []
for state in states:
    if state in east_states:
        # Eastern states: mostly Republican (blue), some random Democrats (red)
        if np.random.rand() < 0.8:
            winner = 'Republican'
            percent = np.random.uniform(52, 70)
        else:
            winner = 'Democrat'
            percent = np.random.uniform(52, 70)
    else:
        # Western states: mostly Democrat (red), some random Republicans (blue)
        if np.random.rand() < 0.8:
            winner = 'Democrat'
            percent = np.random.uniform(52, 70)
        else:
            winner = 'Republican'
            percent = np.random.uniform(52, 70)
    results.append({'state': state, 'abbr': state_abbr[states.index(state)], 'winner': winner, 'percent': percent})

df = pd.DataFrame(results)

# --- Matplotlib/Seaborn horizontal bar "heatmap" ---


plt.figure(figsize=(10, 12))
df_sorted = df.sort_values('percent')
# Create a list of colors for each bar
bar_colors = ['blue' if w == 'Republican' else 'red' for w in df_sorted['winner']]
sns.barplot(y='state', x='percent', data=df_sorted, palette=bar_colors)

plt.title('Simulated Election Results: Republican (Blue) vs Democrat (Red)')
plt.xlabel('Winning Percentage')
plt.ylabel('State')
plt.xlim(50, 75)
plt.tight_layout()
plt.show()

# --- Plotly USA Map Choropleth ---
fig = px.choropleth(
    df,
    locations='abbr',
    locationmode='USA-states',
    color='winner',
    color_discrete_map={'Republican': 'blue', 'Democrat': 'red'},
    hover_name='state',
    hover_data={'percent': True, 'winner': True, 'abbr': False},
    scope='usa',
    title='Simulated Election Results: Republican (Blue) vs Democrat (Red)'
)
fig.update_traces(marker_line_width=0.5, marker_line_color='white')
fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load iris dataset
df = sns.load_dataset('iris')

# Compute statistics for sepal_length
mean = df['sepal_length'].mean()
median = df['sepal_length'].median()
mode = df['sepal_length'].mode()[0]
q1, q2, q3 = df['sepal_length'].quantile([0.25, 0.5, 0.75])
std = df['sepal_length'].std()
var = df['sepal_length'].var()
rng = df['sepal_length'].max() - df['sepal_length'].min()
conf_int = 1.96 * std / np.sqrt(len(df))

# Visualize with a boxplot and overlay statistics
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['sepal_length'], color='lightblue')
plt.axvline(mean, color='red', linestyle='--', label=f'Mean: {mean:.2f}')
plt.axvline(median, color='green', linestyle='-', label=f'Median: {median:.2f}')
plt.axvline(mode, color='purple', linestyle=':', label=f'Mode: {mode:.2f}')
plt.axvspan(mean - conf_int, mean + conf_int, color='orange', alpha=0.2, label='95% CI (mean)')
plt.title('Sepal Length Statistics (Iris Dataset)')
plt.xlabel('Sepal Length (cm)')
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load iris dataset
df = sns.load_dataset('iris')

# Compute statistics for sepal_length
mean = df['sepal_length'].mean()
median = df['sepal_length'].median()
mode = df['sepal_length'].mode()[0]
q1, q2, q3 = df['sepal_length'].quantile([0.25, 0.5, 0.75])
std = df['sepal_length'].std()
var = df['sepal_length'].var()
rng = df['sepal_length'].max() - df['sepal_length'].min()
conf_int = 1.96 * std / np.sqrt(len(df))

# Visualize with a histogram and overlay statistics
plt.figure(figsize=(8, 6))
sns.histplot(df['sepal_length'], bins=20, color='lightblue', kde=True)
plt.axvline(mean, color='red', linestyle='--', label=f'Mean: {mean:.2f}')
plt.axvline(median, color='green', linestyle='-', label=f'Median: {median:.2f}')
plt.axvline(mode, color='purple', linestyle=':', label=f'Mode: {mode:.2f}')
plt.axvspan(mean - conf_int, mean + conf_int, color='orange', alpha=0.2, label='95% CI (mean)')
plt.title('Sepal Length Statistics (Iris Dataset)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
